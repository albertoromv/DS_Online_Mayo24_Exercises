## Introducción a Git y Ramas

### 1. Repositorio de Git

Un "Repo" de Git es un espacio de trabajo que rastrea y gestiona archivos dentro de una carpeta. Cada vez que queremos usar Git con un proyecto, aplicación, etc., necesitamos crear un nuevo repositorio de Git. Podemos tener tantos repositorios en nuestra máquina como sea necesario, cada uno con historias y contenidos separados.


### 2. Flujo de trabajo en GIT


#### La característica más importante de Git

#### Staging Area

- `git add`:Antes de realizar un commit en Git, es necesario agregar los cambios al "staging area". Este es un espacio intermedio donde los cambios se preparan para ser incluidos en el próximo commit. El staging area permite una mayor flexibilidad y control sobre qué cambios se incluirán en cada commit.

#### Commit

- `git commit -m "mensaje"`:En Git, un "commit" es una acción mediante la cual se registran los cambios realizados en los archivos de un repositorio. Representa un punto en la historia del proyecto y almacena una instantánea de los archivos en ese momento específico. Cada commit tiene un identificador único y un mensaje que describe los cambios realizados.

#### Hash en los Commits

- `85a84ef7b8b261ea680308d7f2d672f161e8a01a`:En Git, cada commit está identificado por un hash, una cadena única de caracteres que actúa como una huella digital para ese commit en particular. Este hash permite que Git rastree y gestione cada commit de manera única.


#### Primeros Comandos Git
-  Ir al directorio de trabajo.
- `git status`: Proporciona información sobre el estado actual de un repositorio Git y su contenido.
- `git init`: Se utiliza para iniciar un nuevo repositorio Git en un directorio. Crea un nuevo repositorio vacío y comienza a realizar un seguimiento de los cambios en ese directorio.

## Ramas - Branches


### 4. HEAD en Git

HEAD es un puntero que se refiere a la ubicación actual en tu repositorio. Apunta a una referencia de rama específica y cambia al moverse entre ramas.



### 5. Ramas en Git

Las ramas son líneas de tiempo alternativas para un proyecto en Git. Permiten crear contextos separados para probar nuevas ideas y trabajar en múltiples funcionalidades de manera paralela.

- `git branch <nombre_rama>` creación de rama
- `git switch <nombre_rama>` cambia de rama
- `git checkout <nombre_rama>` cambia de rama
- `git branch -d <nombre_rama>` elimina rama

## Fusión de Ramas o Merge

### 7. Fusión de Ramas (Merge)

Al fusionar ramas en Git, ten en cuenta dos conceptos:

1. Fusionamos ramas, no commits específicos.
2. Siempre fusionamos a la rama HEAD actual.

#### Código en Git Bash para fusionar:

```bash
git checkout <rama_destino>
git merge <rama_fuente>

#### Código para borrar una rama:
```bash
git branch -d <nombre_rama>
git branch -D <nombre_rama>

8. **Resolución de conflictos:**
   - Durante la fusión, pueden surgir conflictos si hay cambios en las mismas líneas de código en ambas ramas. Resuelve estos conflictos manualmente y luego completa la fusión con `git merge --continue`.

9. **Ramificaciones remotas:**
   - Git permite trabajar con ramas remotas, lo que facilita la colaboración en equipos distribuidos. Puedes empujar (`push`) y jalar (`pull`) ramas entre repositorios locales y remotos.

10. **Flujo de trabajo de ramas:**
    - Las ramas son fundamentales en el flujo de trabajo de Git. Permiten a los equipos trabajar en paralelo en diferentes características o correcciones de errores sin interferir entre sí. Esto facilita la colaboración y la gestión de cambios en proyectos grandes.

11. **Etiquetado (tagging):**
    - Además de las ramas, Git permite etiquetar puntos específicos en la historia con "tags". Los tags se utilizan comúnmente para marcar versiones específicas del software, como lanzamientos estables o hitos importantes.

Estos conceptos proporcionan una base sólida para comprender y utilizar eficazmente las ramas en Git en el desarrollo de software.

- **Git** es un sistema de control de versiones distribuido que funciona localmente en la máquina del usuario.
- **GitHub** es una plataforma en línea que utiliza Git para alojar proyectos, facilitando la colaboración y proporcionando herramientas adicionales.

**Importante:** Aunque GitHub es una de las plataformas más populares que utiliza Git, existen otras opciones, como GitLab o Bitbucket, que ofrecen funcionalidades similares. Cada una de estas plataformas tiene sus propias características y ventajas.

➡️ **Ramas Remotas**


Las **"remote branches"** (ramas remotas) en Git son referencias a ramas que existen en repositorios remotos. Un repositorio remoto es una copia de un repositorio Git que se encuentra en otro lugar, como en un servidor en línea o en la máquina de otro colaborador.

Cuando clonas un repositorio remoto o agregas uno como remoto a tu repositorio local, obtienes información sobre las ramas que existen en ese repositorio remoto. Estas ramas en el repositorio remoto se llaman **"remote branches"** y pueden ser referenciadas desde tu repositorio local.


# FETCHING
La operación de *fetch* nos permite descargar cambios desde un repositorio remoto, PERO esos cambios no se integrarán automáticamente en nuestros archivos de trabajo.

Te permite ver en qué han estado trabajando los demás, sin tener que fusionar esos cambios en tu repositorio local.

Piensa en ello como "por favor, ve y obtén la información más reciente de GitHub, pero no arruines mi directorio de trabajo".


# Git FETCH

También podemos obtener información de una rama específica de un remoto utilizando `git fetch <remoto> <rama>`.

Por ejemplo, `git fetch origin master` recuperaría la información más reciente de la rama master en el repositorio remoto llamado origin.

# GITHUB

### Funciones Específicas de GitHub

1. **Interfaz Web:**
   - GitHub ofrece una interfaz web que facilita la visualización y gestión de repositorios, ramas, Pull Requests y otros elementos del proyecto.

2. **Pull Requests (PR):**
   - Aunque el concepto de Pull Request existe en Git, GitHub añade una capa adicional de funcionalidad y facilita la revisión del código, la discusión y la fusión de cambios de manera colaborativa a través de su interfaz web.

3. **Issues y Proyectos:**
   - GitHub permite el seguimiento de problemas y la organización de tareas en tableros de proyectos. Estas funciones no están directamente en Git y son herramientas adicionales para la gestión de proyectos.

4. **GitHub Actions:**
   - GitHub Actions es un servicio de integración continua (CI) y despliegue continuo (CD) integrado directamente en GitHub.

5. **Gestión de Equipos y Organizaciones:**
   - GitHub proporciona herramientas para crear organizaciones y gestionar equipos dentro de esas organizaciones. Estas funciones no son parte de Git, pero son esenciales para la colaboración y la administración de permisos en proyectos más grandes.

6. **GitHub Pages:**
   - GitHub Pages es un servicio que permite alojar sitios web directamente desde tus repositorios en GitHub.

7. **GitHub Gist:**
   - Gist es una característica de GitHub que permite compartir fragmentos de código, notas y archivos de manera rápida y sencilla.

8. **GitHub Marketplace:**
   - La Marketplace de GitHub ofrece una variedad de aplicaciones y herramientas integrables que pueden mejorar las capacidades y la funcionalidad de tu repositorio o proyecto.

9. **GitHub Discussions:**
   - GitHub Discussions proporciona un espacio para la comunicación y discusiones en torno a un proyecto.

Estas funciones adicionales hacen de GitHub una plataforma completa para el desarrollo colaborativo de software, con herramientas integradas para facilitar la colaboración, la gestión de proyectos y la automatización de flujos de trabajo.
